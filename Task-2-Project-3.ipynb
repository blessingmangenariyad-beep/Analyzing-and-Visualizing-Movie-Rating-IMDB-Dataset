```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv(r"D:\internship\Datasets\IMDB-Movie-Data.csv")
```

```python
#Dropping missing Values
df.dropna(axis=0)
```

```python
#Check for duplicates
df.duplicated().sum()
```

```python
#All statistics about the Dataset including categorical data
df.describe(include='all')
```

```python
#Summary Statistics of Ratings
mean_rating = df['Rating'].mean()
median_rating = df['Rating'].median()
mode_rating = df['Rating'].mode()[0]

print("Mean Rating:", round(mean_rating, 2))
print("Median Rating:", median_rating)
print("Mode Rating:", mode_rating)
```

```python
df.groupby('Year')['Votes'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,5))

sns.barplot(x='Year', y='Votes', data=df)

plt.xlabel('Year')
plt.ylabel('Votes')
plt.title('Votes By Year')

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Rating'], bins=20, kde=True)
plt.title("Distribution of Movie Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Rating'])
plt.title("Box Plot of Movie Ratings")
plt.show()
```

```python
#Top Rated Movies
top_movies = df.sort_values(by='Rating', ascending=False).head(10)
top_movies[['Title', 'Rating']]
```

```python
top_Director = df.sort_values(by='Rating', ascending=False).head(10)
top_Director[['Director', 'Rating']]
```

```python
#Top Rated Genres
genre_ratings = df.groupby('Genre')['Rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
genre_ratings.head(10).plot(kind='bar')
plt.title("Top Rated Genres")
plt.xlabel("Genre")
plt.ylabel("Average Rating")
plt.show()
```

```python
sns.countplot(x ='Year', data=df)
plt.title("Number of movies per Year")
```

```python
#Classifying Movies 
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
    
```

```python
plt.figure(figsize=(14,10))

# Histogram
plt.subplot(2,2,1)
sns.histplot(df['Rating'], bins=20, kde=True, color='skyblue')
plt.title("Rating Distribution")

# Box Plot
plt.subplot(2,2,2)
sns.boxplot(x=df['Rating'])
plt.title("Ratings Box Plot")

# Top Genres
plt.subplot(2,2,3)
genre_ratings.head(5).plot(kind='bar')
plt.title("Top 5 Genres")

# Top Movies
top_movies = top_movies.sort_values(by='Rating', ascending=True)

plt.subplot(2,2,4)
sns.barplot(
    x='Rating',
    y='Title',
    data=top_movies,
    palette='tab10'
)



plt.xlabel("Rating")
plt.ylabel("Movie Title")
plt.Title("Top Rated Movies")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,10))

# 1 Histogram
plt.subplot(2,2,1)
sns.histplot(df['Rating'], bins=20, kde=True, color='cornflowerblue')
plt.title("Rating Distribution")

# 2 Box Plot
plt.subplot(2,2,2)
sns.boxplot(x=df['Rating'], color='lightcoral')
plt.title("Ratings Box Plot")

# 3 Top Genres
plt.subplot(2,2,3)
sns.barplot(
    x=genre_ratings.head(5).index,
    y=genre_ratings.head(5).values,
    palette='viridis'
)
plt.title("Top 5 Genres")
plt.xticks(rotation=30)

# 4 Top Movies
plt.subplot(2,2,4)
sns.barplot(
    x='Rating',
    y='Title',
    data=top_movies,
    palette='magma'
)
plt.title("Top Rated Movies")

plt.tight_layout()
plt.show()
```
